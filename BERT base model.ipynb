```python
import tensorflow_hub as hub

import tensorflow_text as text
```

```python
preprocessed_url = "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"

encoder_url="https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4"
```

```python
bert_preprocess_model = hub.KerasLayer(preprocessed_url)
```

```python
sample_text=["Every one" , "its a fine day","All is well", "everthing looks good"]
```

```python
sample_text_preprocessed = bert_preprocess_model(sample_text)
```

```python
bert_model = hub.KerasLayer(encoder_url)
bert_enc_model_pre_prc = bert_model(sample_text_preprocessed)
```
